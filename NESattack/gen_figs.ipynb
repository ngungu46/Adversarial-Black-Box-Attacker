```python
import os 
import matplotlib.pyplot as plt 
import json 
from pprint import pprint  


stats = [os.path.join("outputs", "run_0", fpath) for fpath in os.listdir(os.path.join("outputs", "run_0")) if fpath.endswith(".json")]
```

```python
queries = []

for stat in stats: 
    with open(stat) as f: 
        data = json.load(f) 
        # pprint(data)
        
        if data["status"] == "SUCCESS": 
            queries.append(data["num_queries_needed"])
        
```

```python
import numpy as np 

np.median(queries)
```

```python
plt.hist(queries, edgecolor='black', linewidth=1.2)
plt.xlabel("Number of Queries")
plt.ylabel("Count")
plt.savefig("hist_nes_imagenet.png")
# plt.title("")
```
